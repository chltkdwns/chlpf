```python
import os

# 경로 설정
train_dir = '/content/all/train'
test_dir = '/content/all/test'

# 훈련 데이터 클래스 수 확인
train_classes = os.listdir(train_dir)
print("훈련 데이터 클래스 수:")
for cls in train_classes:
    cls_path = os.path.join(train_dir, cls)
    print(f"{cls}: {len(os.listdir(cls_path))}개")

# 검증 데이터 클래스 수 확인
test_classes = os.listdir(test_dir)
print("\n검증 데이터 클래스 수:")
for cls in test_classes:
    cls_path = os.path.join(test_dir, cls)
    print(f"{cls}: {len(os.listdir(cls_path))}개")
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import random

# Seed 고정
seed_value = 42
np.random.seed(seed_value)
tf.random.set_seed(seed_value)
random.seed(seed_value)

# 데이터 증강 설정
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.15,
    zoom_range=0.15,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    '/content/all/train',  # 경로 확인 필요
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    '/content/all/test',  # 경로 확인 필요
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

# 사전 훈련된 모델 로드 및 레이어 구성
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(150, 150, 3))
base_model.trainable = False  # 미세 조정을 위해 가중치 고정

# 모델 구성
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dropout(0.4),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# 모델 컴파일
model.compile(optimizer=Adam(learning_rate=1e-4),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# 콜백 설정 (조기 종료 없이 ReduceLROnPlateau만 사용)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=3, min_lr=1e-6)

# EarlyStopping 콜백 설정: 'val_loss'가 5 에포크 동안 개선되지 않으면 학습 중단
early_stopping = EarlyStopping(
    monitor='val_loss',  # 검증 손실 기준
    patience=5,          # 성능 개선 대기 횟수
    restore_best_weights=True  # 최적의 가중치를 복원
)

# 모델 학습
history = model.fit(
    train_generator,
    epochs=100,  # 최대 에포크 수를 충분히 길게 설정
    validation_data=test_generator,  # validation_generator를 test_generator로 변경
    callbacks=[early_stopping]  # 조기 종료 콜백 추가
)

# 훈련 결과 요약
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']

avg_train_accuracy = np.mean(train_accuracy)
avg_val_accuracy = np.mean(val_accuracy)
avg_train_loss = np.mean(train_loss)
avg_val_loss = np.mean(val_loss)

print(f"평균 훈련 정확도: {avg_train_accuracy:.4f}")
print(f"평균 검증 정확도: {avg_val_accuracy:.4f}")
print(f"평균 훈련 손실: {avg_train_loss:.4f}")
print(f"평균 검증 손실: {avg_val_loss:.4f}")

# 훈련 및 검증 손실, 정확도 시각화
epochs_range = range(1, len(train_accuracy) + 1)

plt.figure(figsize=(12, 5))

# 정확도 그래프
plt.subplot(1, 2, 1)
plt.plot(epochs_range, train_accuracy, label='훈련 정확도')
plt.plot(epochs_range, val_accuracy, label='검증 정확도')
plt.title('훈련 및 검증 정확도')
plt.xlabel('에포크')
plt.ylabel('정확도')
plt.legend()

# 손실 그래프
plt.subplot(1, 2, 2)
plt.plot(epochs_range, train_loss, label='훈련 손실')
plt.plot(epochs_range, val_loss, label='검증 손실')
plt.title('훈련 및 검증 손실')
plt.xlabel('에포크')
plt.ylabel('손실')
plt.legend()

plt.show()
```
